```python
def mape_rmspe(model,X_train,X_test,y_train,y_test,n_runs=100):
    mape = 0
    rmspe = 0
    for i in range(n_runs):
        model.fit(X_train,y_train)
        pred = model.predict(X_test)
        pred = pd.DataFrame(pred,index=y_test.index)
        pred.columns =['Predictions']
        bench = pd.concat([pred,y_test],axis=1)
        bench['% of error'] = (bench['Predictions'] - bench[bench.columns[1]])/(bench[bench.columns[1]]/100)
        mape += bench['% of error'].abs().mean()
        rmspe += ((bench['% of error']**2).mean())**0.5
    mape /= n_runs
    rmspe /= n_runs
    return mape,rmspe
```

```python
random_forest_macro_mape,random_forest_macro_rmspe = mape_rmspe(black_box_model,X_train,X_test,y_train,y_test)
print('Random Forest - Macro Economics')
print('Mean Absolute Percentage Error:')
print(random_forest_macro_mape)
print('Root Mean Squared Percentage Error:')
print(random_forest_macro_rmspe)
```
